```python
s=int(input("enter the number of student"))
st=1
while(st<=s):
    su=int(input("enter the number of subject"))
    stt=1
    tot=0
    avg=0
    while(stt<=su):
        mk=int(input("enter the mark"))
        tot+=mk
        avg=tot/su
        stt+=1
    print("total "+str(tot)+"\n"+"avarage "+str(avg))
    st+=1
print()
for i in range(1,s+1):
    sub=int(input("enter the number of subject"))
    tot=0
    avg=0
    for j in range(1,sub+1):
        mrk=int(input("enter the mark in "+str(j)+"-th subject"))
        tot+=mrk
        avg=tot/sub
print("total "+str(tot)+"\navarage "+str(avg))

    
```

```python
n=int(input("enter the number of rount"))
print()
score=0
s=1
while(s<=n):
    r=int(input("enter the number of run"))
    if r<=6 and r>=0:
        score+=r
        s+=1
    else:
        print("invalid runs")
print(score)
print()
sc=0
for i in range(1,n+1):
    r=int(input("enter the number of run"))
    if r<=6 and r>=0:
        sc+=r
    else:
        print("invalid runs")
print("total score "+str(sc))
```

```python
print("1.addition \n2.subraction \n3.multiplicatin \n4.division")
while(True):
    ch=input("enter your choice")
    n1=int(input("enter the first number"))
    n2=int(input("enter the second number"))
    if(ch=='1'):
        print(n1+n2)
    elif(ch=='2'):
        print(n1-n2)
    elif(ch=='3'):
        print(n1*n2)
    elif(ch=='4'):
        print(n1/n2)
    else:
        print("enter the valid choice")
    ap=input("do you want to continue \n1.yes \n2.no\n")
    if ap=='2':
        break
        
```

```python
import random
while(True):
    nr=int(input("enter the number of round you want"))
    print("1.rock 2.paper 3.scissor")
    cwin=0
    pwin=0
    if nr%2==0:
        print("enter odd number of rounds")
    else:
        s=1
        while(s<=nr):
            cch=random.randint(1,3)
            pch=int(input("enter your choice"))
            print("computer choice "+str(cch))
            if pch>=1 and pch<=3:
                if cch==pch:
                    print("draw")
                    cwin+=1
                    pwin+=1
                elif cch==1 and pch==3 or cch==2 and pch==1 or cch==3 and pch==2:
                    print("computer win")
                    cwin+=1
                else:
                    print("player win")
                    pwin+=1
                s+=1
                if(cwin>nr//2):
                    print("----------------------computer win--------------------------")
                    break
                elif(pwin>nr//2):
                    print("------------------------player win------------------------")
                    break
            else:
                print("invalid choice")
    ans=input("do you want to continue \n1.yes\n2.no\n")
    if ans=='2':
        break
```

```python
import random
print("-----------guess the number---------------")
print()
r=random.randint(1,100)
c=0
while(True):
    uch=int(input("enter your guessing"))
    if(uch==r):
        print("correct")
        break
    elif uch>r:
        print("your number is greater guess lower number")
    else:
        print("your number is lower guess higher number")
    c+=1
print("total number of guesses "+str(c))
```
