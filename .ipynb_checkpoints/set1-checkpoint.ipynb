```python
import math
p=int(input("enter the principle amount"))
n=int(input("enter the number of number of time interest is compounded  per time period"))
r=int(input("rate of interest"))
t=int(input("number of years"))
r=r/100;
compound=p*math.pow((1+r/n),(n*t))
interest=compound-p
print("final amount "+str(compound)+" and compound interest "+str(interest))
```
